```python
import graphviz
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head(3)
```

```python
df.info()
```

```python
y = df['output']
```

```python
x = df.drop('output',axis=1)
```

```python
tree = DecisionTreeClassifier()
```

```python
model = tree.fit(x,y)
```

```python
model.score(x,y)
```

```python
df.info()
```

```python
#train_test_split ile
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 31 ,train_size = 0.7)
```

```python
tree = DecisionTreeClassifier()
```

```python
model = tree.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[31,1,2,130,240,0,0,150,0,2,0,0,2]])
```

```python
dot = export_graphviz(model,feature_names = x.columns , filled = True)
```

```python
gorsel = graphviz.Source(dot)
```
